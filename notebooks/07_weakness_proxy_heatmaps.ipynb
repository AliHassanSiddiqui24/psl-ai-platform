```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/processed/ball_by_ball/psl_ball_by_ball_clean.csv")
dismissal_df = df[(df["wicket"] == 1) & (df["player_out"].notna())].copy()
dismissal_df.head()
```

```python
player_name = "Babar Azam"
player_df = dismissal_df[dismissal_df["player_out"] == player_name].copy()
player_df.head()
```

```python
phase_wicket = pd.pivot_table(
    player_df,
    index="phase",
    columns="wicket_type",
    values="match_id",
    aggfunc="count",
    fill_value=0,
)
phase_wicket
```

```python
# Check karein ke kitne records mile
print(f"Total dismissals found for {player_name}: {len(player_df)}")

# Agar data khali hai, to unique names check karein ke file mein kya naam likha hai
if len(player_df) == 0:
    print("Available players in dismissal data:")
    print(dismissal_df["player_out"].unique()[:10]) # Pehle 10 naam check karein
```

```python
if not phase_wicket.empty:
    plt.figure(figsize=(10, 6))
    sns.heatmap(phase_wicket, annot=True, fmt="g", cmap="Reds")
    plt.title(f"{player_name} — Phase vs Wicket Type")
    plt.show()
else:
    print(f"Oops! {player_name} ka koi dismissal record nahi mila, isliye graph nahi ban sakta.")
```
