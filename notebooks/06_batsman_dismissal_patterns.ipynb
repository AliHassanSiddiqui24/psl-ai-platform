```python
import pandas as pd
from pathlib import Path
```

```python
df = pd.read_csv("data/processed/ball_by_ball/psl_ball_by_ball_clean.csv")
df.head()
```

```python
dismissal_df = df[(df["wicket"] == 1) & (df["player_out"].notna())].copy()
dismissal_df.head()
```

```python
dismissal_df["wicket_type"].value_counts()
```

```python
dismissal_df["player_out"].value_counts().head(20)
```

```python
player_name = "Babar Azam"
dismissal_df[dismissal_df["player_out"] == player_name]["wicket_type"].value_counts()
```

```python
summary = (
    dismissal_df.groupby(["player_out", "wicket_type"])
    .size()
    .reset_index(name="dismissal_count")
)
summary.head(20)
```

```python
top_batters = dismissal_df["player_out"].value_counts().head(10).index.tolist()
summary[summary["player_out"].isin(top_batters)].head(30)
```
